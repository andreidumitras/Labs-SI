```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-20, 20, 100)
# print(x)
y = x**2*np.sin(x)
plt.plot(x, y, 'o-')
plt.title('Functia de gradul 1')
plt.xlabel('timp')
plt.ylabel('f(timp)')
plt.grid()
plt.show()
```
